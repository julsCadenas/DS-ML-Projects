```python
# Importing pandas, numpy and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
pd.pandas.set_option('display.max_columns',None)
netflix_df = pd.read_csv("../data/netflix_data.csv")
```

### **1. Filter the data for movies released in the 1990s.**
- You'll want to subset the DataFrame to keep only the movies and then filter the years so that you are working with rows where the release year is between 1990 and 1999.

```python
netflix_movies = netflix_df[netflix_df['type'] == 'Movie']
```

```python
netflix1990s = netflix_movies[(netflix_movies['release_year'] < 2000) & (netflix_movies['release_year'] >= 1990)]
netflix1990s
```

### **2. Find the most frequent movie duration**
- Visualize the data to view the distributions of durations for movies released in the 1990s.

```python
netflix_movies
```

```python
def duration_hist(df, bin_count):
    plt.hist(df['duration'], edgecolor='black', bins=bin_count)
    plt.xlabel('Duration')
    plt.ylabel('Frequency')
    plt.title('Distribution of Movie Durations')
    plt.tight_layout()
    plt.show()

bin_count = int(np.sqrt(len(netflix1990s['duration'])))
duration_hist(netflix1990s, bin_count)
```

```python
duration = netflix1990s['duration'].mode()[0]
duration
```

### **3. Count the number of short action movies from the 1990s.**
- First, subset the data again to keep only movies with the correct genre. Then, iterate through the data and count the number of movies depending on the condition of a duration that is less than 90.

```python
genres = netflix1990s['genre'].unique()
print(genres)

action = netflix1990s[netflix1990s['genre'] == 'Action']
action.shape
```

```python
short_90s_action = action['duration'] <= 90
count = short_90s_action.sum()

count
```
